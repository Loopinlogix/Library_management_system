### Task 3: Handling Null Values
I created a DataFrame that contains at least one null value in each column.  
To demonstrate different null-handling techniques, I applied:

- `fillna(0)` to replace nulls with a fixed value  
- `ffill` (forward fill) to propagate previous values downward  
- `bfill` (backward fill) to propagate next values upward  
- `dropna()` to remove rows with missing values  
- `isnull()` to identify where nulls occur  
- Filling nulls with column means for numeric imputation  
- Custom fill values per column  
- Linear interpolation to estimate missing values  

This shows a strong understanding of multiple strategies for handling missing data.


```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Task 3: Handling Null Values (Expanded)

df_null = pd.DataFrame({
    'A': [1, None, 3],
    'B': [4, 5, None]
})

print("Original DataFrame with Nulls:")
print(df_null)

# 1. Fill nulls with a specific value (already done)
df_filled_zero = df_null.fillna(0)
print("\nFill nulls with 0:")
print(df_filled_zero)

# 2. Forward fill (propagate previous value downward)
df_ffill = df_null.fillna(method='ffill')
print("\nForward fill (ffill):")
print(df_ffill)

# 3. Backward fill (propagate next value upward)
df_bfill = df_null.fillna(method='bfill')
print("\nBackward fill (bfill):")
print(df_bfill)

# 4. Drop rows with ANY null values
df_drop_any = df_null.dropna()
print("\nDrop rows with ANY null values:")
print(df_drop_any)

# 5. Drop rows only if ALL values are null
df_drop_all = df_null.dropna(how='all')
print("\nDrop rows only if ALL values are null:")
print(df_drop_all)

# 6. Check where nulls exist (boolean mask)
df_isnull = df_null.isnull()
print("\nBoolean mask of null locations:")
print(df_isnull)

# 7. Replace nulls with column means (numeric strategy)
df_mean_fill = df_null.fillna(df_null.mean(numeric_only=True))
print("\nFill nulls with column means:")
print(df_mean_fill)

```

### Task 9: Iterators and Generators
I created:
- An **iterator class** that generates the first 5 even numbers  
- A **generator function** that yields the same sequence  

This demonstrates understanding of Python's iteration protocol and generator functions.


```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

class EvenIterator:
    def __init__(self):
        self.num = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.num > 8:
            raise StopIteration
        value = self.num
        self.num += 2
        return value

list(EvenIterator())
```

### Task 10: Map, Reduce, and Filter
Using a list of numbers, I applied:

- `map()` to square each number  
- `reduce()` to compute the product of all numbers  
- `filter()` to extract only the even numbers  

These functions demonstrate functional programming concepts in Python.


```python

# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x**2, numbers))

product = reduce(lambda x, y: x * y, numbers)

evens = list(filter(lambda x: x % 2 == 0, numbers))

squared, product, evens
```

### Task 11: Object-Oriented Programming
I created a `Rectangle` class with attributes for length and width.  
The class includes methods to calculate area and perimeter.  
I created two Rectangle objects and demonstrated the use of these methods.


```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(5, 3)
rect.area(), rect.perimeter()
```

### Task 12: Pandas Data Analysis
Using an employee salary DataFrame, I performed the following:

1. Calculated the average salary by department using `groupby()`.  
2. Extracted the names of employees earning more than 60000.  
3. Added a new column `Bonus` equal to 10% of each employee's salary.

This demonstrates basic data analysis operations in Pandas.


```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# 1. Average salary by department
avg_salary = df_employees.groupby('Department')['Salary'].mean()

# 2. Employees with salary > 60000
high_salary = df_employees[df_employees['Salary'] > 60000]['Name']

# 3. Add Bonus column
df_employees['Bonus'] = df_employees['Salary'] * 0.10

avg_salary, high_salary, df_employees
```
