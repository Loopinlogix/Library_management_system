##Functions and Conditional Statements
Creating a reuseable check_availability() to verify if a book exists in the library. This would be showing functions like definition, parameters, returns values and conditional logic.

```python
# 5. Functions and Conditional Statements

def check_availability(book_name):
    """
    Check if a book is available in the library.

    Args:
        book_name (str): The title of the book to check

    Returns:
        bool: True if the book is in the list, False otherwise
    """
    # Convert to lowercase for case-insensitive comparison

    book_name_lower = book_name.lower().strip()

    # Check if the book exists in the list
    # Using a generator expression with any() for case-insensitive check
    if any(book.lower() == book_name_lower for book in books):
        return True
    else:
        return False

# Test the function with various cases
print("Testing check_availability() function:")
print("-" * 50)

# Test 1: Book that exists
test_book = "Python Basics"
print(f"Is '{test_book}' available? {check_availability(test_book)}")

# Test 2: Book that doesn't exist
test_book2 = "Java Programming"
print(f"Is '{test_book2}' available? {check_availability(test_book2)}")

# Test 3: Case insensitive check
test_book3 = "python basics"
print(f"Is '{test_book3}' available? (case-insensitive) {check_availability(test_book3)}")

# Test 4: Empty string (edge case)
print(f"Is '' available? {check_availability('')}")

# Test 5: Book with extra whitespace
test_book4 = "  Python Basics  "
print(f"Is '{test_book4.strip()}' with whitespace available? {check_availability(test_book4)}")
```

##File I/O
File Operations - writing to a file and reading from a file which inclues error handling for common file operations issues.

```python
# 6. File I/O

def save_books_to_file(filename):
    """
    Save the list of books to a text file.

    Args:
        filename (str): The name of the file to save to
    """
    try:
        # 'w' mode opens the file for writing (creates if doesn't exist, overwrites if exists)
        with open(filename, 'w') as f:
            for book in books:
                f.write(f"{book}\n")
        print(f"Books successfully saved to {filename}")
    except PermissionError:
        print(f"Error: Permission denied. Cannot write to {filename}")
    except Exception as e:
        print(f"Error saving books: {e}")

# TODO: Implement a function to read books from a file
def read_books_from_file(filename):
    """
    Read books from a text file and return them as a list.

    Args:
        filename (str): The name of the file to read from

    Returns:
        list: A list of book titles, or empty list if file not found
    """
    book_list = []
    try:
        # Check if file exists first
        if not os.path.exists(filename):
            print(f"Error: File '{filename}' does not exist.")
            return book_list

        # 'r' mode opens the file for reading
        with open(filename, 'r') as f:
            # readlines() returns a list of lines, strip() removes newline characters
            book_list = [line.strip() for line in f.readlines() if line.strip()]

        print(f"Successfully read {len(book_list)} books from {filename}")
        return book_list

    except PermissionError:
        print(f"Error: Permission denied. Cannot read {filename}")
        return book_list
    except Exception as e:
        print(f"Error reading books: {e}")
        return book_list

# Test file I/O
print("Testing File I/O Operations:")
print("=" * 50)

# Save books to file
save_books_to_file("books.txt")

# Display file contents using shell command (Colab specific)
print("\nContents of books.txt:")
print("-" * 30)
!cat books.txt

# Read books from file
print("\nReading books from file:")
print("-" * 30)
loaded_books = read_books_from_file("books.txt")
for i, book in enumerate(loaded_books):
    print(f"  [{i}] {book}")

# Test error handling: try to read a non-existent file
print("\nTesting error handling (non-existent file):")
print("-" * 30)
read_books_from_file("nonexistent_file.txt")
```

##Pickle for Serialization
Using Python pickle module to save and load the books list. Pickle preserves python object exactly the way they are incl. data types.


```python
# 7. Pickle for serialization

def save_books_pickle(filename):
    """
    Save the books list to a file using pickle serialization.

    Args:
        filename (str): The name of the file to save to
    """
    try:
        # 'wb' mode = write binary (required for pickle)
        with open(filename, 'wb') as f:
            pickle.dump(books, f)
        print(f"Books successfully saved to {filename} (pickle format)")
    except Exception as e:
        print(f"Error saving with pickle: {e}")

# TODO: Implement a function to load books using pickle
def load_books_pickle(filename):
    """
    Load books from a pickle file.

    Args:
        filename (str): The name of the pickle file to load

    Returns:
        list: The loaded books list, or empty list if error occurs
    """
    try:
        # Check if file exists
        if not os.path.exists(filename):
            print(f"Error: Pickle file '{filename}' does not exist.")
            return []

        # 'rb' mode = read binary (required for pickle)
        with open(filename, 'rb') as f:
            loaded_books = pickle.load(f)

        print(f"Successfully loaded {len(loaded_books)} books from {filename}")
        return loaded_books

    except pickle.UnpicklingError:
        print(f"Error: File '{filename}' is not a valid pickle file.")
        return []
    except Exception as e:
        print(f"Error loading with pickle: {e}")
        return []

# Test pickle functions
print("Testing Pickle Serialization:")
print("=" * 50)

# Save books using pickle
save_books_pickle("books.pkl")

# Load books using pickle
print("\nLoading books from pickle:")
print("-" * 30)
loaded_from_pickle = load_books_pickle("books.pkl")
print("Books loaded from pickle:", loaded_from_pickle)

# Verify that the loaded data matches the original
print(f"\nData integrity check: {loaded_from_pickle == books}")

# Test error handling
print("\nTesting error handling (non-existent pickle file):")
print("-" * 30)
load_books_pickle("nonexistent.pkl")
```

##SQlite Database Interation
Using SQlite to store book data in a relational database. Create a table, insert books, and handle database connection correctly.

```python
# 8. SQLite Database Interaction

def create_books_table():
    """
    Create the books table in the SQLite database if it doesn't exist.
    Uses a context manager to ensure the connection is properly closed.
    """
    try:
        # Connect to the database (creates it if it doesn't exist)
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()

        # Create table with id, title, and quantity columns
        cursor.execute('''CREATE TABLE IF NOT EXISTS books
                          (id INTEGER PRIMARY KEY AUTOINCREMENT,
                           title TEXT NOT NULL,
                           quantity INTEGER DEFAULT 0)''')

        conn.commit()
        print("Books table created successfully (or already exists).")
    except sqlite3.Error as e:
        print(f"Database error creating table: {e}")
    finally:
        if conn:
            conn.close()

# TODO: Implement a function to add a book to the SQLite database
def add_book_to_db(title, quantity):
    """
    Add a book to the SQLite database.

    Args:
        title (str): The title of the book
        quantity (int): The number of copies available
    """
    conn = None
    try:
        # Validate inputs
        if not title or not isinstance(title, str):
            print("Error: Title must be a non-empty string.")
            return

        if not isinstance(quantity, int) or quantity < 0:
            print("Error: Quantity must be a non-negative integer.")
            return

        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()

        # Insert the book into the database
        cursor.execute("INSERT INTO books (title, quantity) VALUES (?, ?)",
                       (title, quantity))

        conn.commit()
        print(f"Successfully added '{title}' with quantity {quantity} to the database.")

        # Return the ID of the inserted book
        return cursor.lastrowid

    except sqlite3.IntegrityError as e:
        print(f"Integrity error: {e}")
    except sqlite3.Error as e:
        print(f"Database error: {e}")
    finally:
        if conn:
            conn.close()

def get_all_books_from_db():
    """
    Retrieve all books from the database.
    """
    conn = None
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM books")
        rows = cursor.fetchall()
        return rows
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        return []
    finally:
        if conn:
            conn.close()

# Test database functions
print("Testing SQLite Database Operations:")
print("=" * 50)

# Create the table
create_books_table()

# Add books to the database
print("\nAdding books to database:")
print("-" * 30)
add_book_to_db("Python for Beginners", 10)
add_book_to_db("Advanced Python", 5)
add_book_to_db("Data Science with Python", 8)

# Test error handling with invalid inputs
print("\nTesting error handling:")
print("-" * 30)
add_book_to_db("", 5)  # Empty title
add_book_to_db("Test Book", -1)  # Negative quantity

# Display all books in the database
print("\nAll books in the database:")
print("-" * 30)
all_books = get_all_books_from_db()
print(f"{'ID':<5} {'Title':<30} {'Quantity':<10}")
print("-" * 50)
for book in all_books:
    print(f"{book[0]:<5} {book[1]:<30} {book[2]:<10}")
```

##Web API (Github API)
Fetching data from the Github API to display the 5 top most starred python libraries.  This will show how to making HTTP GET requests and paring JSON responses.

```python
# Web API Interaction (GitHub API)

def get_python_repos():
    """
    Fetch the top 5 most starred Python repositories from GitHub.

    Returns:
        list: Names of the top 5 Python repositories
    """
    url = "https://api.github.com/search/repositories"
    params = {
        "q": "language:python",
        "sort": "stars",
        "order": "desc"
    }

    try:
        # Make the GET request to the GitHub API
        response = requests.get(url, params=params, timeout=10)

        # Check if the request was successful
        response.raise_for_status()

        # Parse the JSON response
        data = response.json()

        # Extract repository names (top 5)
        repos = [repo['name'] for repo in data['items'][:5]]

        # Also extract star counts for additional info
        repo_details = [(repo['name'], repo['stargazers_count'], repo['html_url'])
                        for repo in data['items'][:5]]

        return repos, repo_details

    except requests.exceptions.Timeout:
        print("Error: Request timed out. Please try again later.")
        return [], []
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e}")
        return [], []
    except requests.exceptions.RequestException as e:
        print(f"Request Error: {e}")
        return [], []
    except (KeyError, ValueError) as e:
        print(f"Data parsing error: {e}")
        return [], []

# Test API function
print("Fetching Top Python Repositories from GitHub:")
print("=" * 50)

repo_names, repo_details = get_python_repos()

if repo_names:
    print("\nTop 5 Python repositories on GitHub:")
    print("-" * 50)
    for i, (name, stars, url) in enumerate(repo_details, 1):
        print(f"{i}. {name}")
        print(f"   Stars: {stars:,}")
        print(f"   URL: {url}")
        print()
else:
    print("Failed to fetch repositories.")
```

## Add on Features
Additional functions added to the Library Management system, removeing old books, update book information, and a simple command line interface.


```python
# 10. Bonus Features

def remove_book(book_name):
    """
    Remove a book from the library list.

    Args:
        book_name (str): The title of the book to remove

    Returns:
        bool: True if removed successfully, False otherwise
    """
    global books
    book_name_lower = book_name.lower().strip()

    # Find the book (case-insensitive)
    for book in books:
        if book.lower() == book_name_lower:
            books.remove(book)
            print(f"Successfully removed '{book}' from the library.")
            return True

    print(f"Book '{book_name}' not found in the library.")
    return False

def update_book(old_title, new_title):
    """
    Update the title of a book in the library.

    Args:
        old_title (str): The current title of the book
        new_title (str): The new title to update to

    Returns:
        bool: True if updated successfully, False otherwise
    """
    global books
    old_title_lower = old_title.lower().strip()

    if not new_title.strip():
        print("Error: New title cannot be empty.")
        return False

    for i, book in enumerate(books):
        if book.lower() == old_title_lower:
            books[i] = new_title
            print(f"Successfully updated '{old_title}' to '{new_title}'.")
            return True

    print(f"Book '{old_title}' not found in the library.")
    return False

def display_library_stats():
    """
    Display statistics about the library.
    """
    print("\n" + "=" * 50)
    print("LIBRARY STATISTICS")
    print("=" * 50)
    print(f"Total books: {len(books)}")
    print(f"\nBook titles:")
    for i, book in enumerate(books, 1):
        print(f"  {i}. {book}")
    print("=" * 50)

# Test bonus features
print("Testing Bonus Features:")
print("=" * 50)

# Display current library
display_library_stats()

# Test removing a book
print("\n--- Testing remove_book() ---")
remove_book("Advanced Python")
remove_book("Non-existent Book")  # Test removing a book that doesn't exist

# Test updating a book
print("\n--- Testing update_book() ---")
update_book("Python Basics", "Python Basics - 3rd Edition")
update_book("Non-existent", "Something")  # Test updating a book that doesn't exist

# Display updated library
display_library_stats()
```

##Simple Command Line Interface
a menu driven for user interaction with LMS

```python
# 11. Simple Command-Line Interface

def library_cli():
    """
    Simple command-line interface for the Library Management System.
    """
    global books # Moved global declaration to the top of the function
    while True:
        print("\n" + "=" * 50)
        print("LIBRARY MANAGEMENT SYSTEM")
        print("=" * 50)
        print("1. View all books")
        print("2. Add a book")
        print("3. Check book availability")
        print("4. Remove a book")
        print("5. Update a book")
        print("6. Save to file")
        print("7. Load from file")
        print("8. Exit")
        print("=" * 50)

        choice = input("Enter your choice (1-8): ").strip()

        if choice == '1':
            display_library_stats()

        elif choice == '2':
            title = input("Enter book title: ").strip()
            if title:
                books.append(title)
                print(f"'{title}' added successfully!")
            else:
                print("Title cannot be empty!")

        elif choice == '3':
            title = input("Enter book title to check: ").strip()
            if check_availability(title):
                print(f"'{title}' is available in the library.")
            else:
                print(f"'{title}' is NOT available in the library.")

        elif choice == '4':
            title = input("Enter book title to remove: ").strip()
            remove_book(title)

        elif choice == '5':
            old = input("Enter current title: ").strip()
            new = input("Enter new title: ").strip()
            update_book(old, new)

        elif choice == '6':
            save_books_to_file("books.txt")
            print("Books saved to books.txt")

        elif choice == '7':
            # global books # This is now redundant and can be removed
            loaded = read_books_from_file("books.txt")
            if loaded:
                books = loaded
                print("Books loaded from books.txt")

        elif choice == '8':
            print("Thank you for using the Library Management System!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 8.")

# Uncomment the line below to run the interactive CLI
# library_cli()

print("CLI function defined. Uncomment library_cli() to run it interactively.")
print("\nLibrary Management System operations completed successfully!")
```
